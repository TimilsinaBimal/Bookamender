```python
ratings_ind = {
    'book_id' : [],
    'rating_count' : [],
    'average_rating' : [],
    'ratings_1' : [],
    'ratings_2' : [],
    'ratings_3' : [],
    'ratings_4' : [],
    'ratings_5' : []
}
```

```python
for i in book_ids:
    rating_i = rating_df.loc[rating_df['book_id'] == i]
    ratings_ind['book_id'].append(i)
    rating_count = rating_i.shape[0]
    ratings_ind['rating_count'].append(rating_count)
    total_rating = 0
    for j in range(1,6):
        count = rating_i[rating_i['rating'] == j ].shape[0]
        ratings_ind['ratings_'+str(j)].append(count)
        total_rating += (j * count)
    ratings_ind['average_rating'].append(round((total_rating/rating_count),2))
```
